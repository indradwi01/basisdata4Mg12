# Tabel Transaksi

```python
!mysql -u root -D Toko_Matematika -e "CREATE TABLE Transaksi (id_transaksi INT PRIMARY KEY AUTO_INCREMENT, id_produk INT, jumlah INT, tanggal DATE, FOREIGN KEY (id_produk) REFERENCES Produk(id_produk));"
```

# Tabel Total Penjualan Produk

```python
!mysql -u root -D Toko_Matematika -e "SELECT P.nama_produk, SUM(T.jumlah) AS total_terjual FROM Transaksi T JOIN Produk P ON T.id_produk = P.id_produk GROUP BY P.nama_produk;"
```
